A) Deliverable 1. Create a simple random graph.

```python
import networkx as nx
import matplotlib.pyplot as plt
import random

#n is the number of nodes and p is the probability of having an edge. For this simple graph program I choose ten nodes with 0.75 chance of having an edge.
n = 10
p = 0.75

#create graph. Use for loop to add nodes. 
Graph1 = nx.Graph()
for i in range(n):
  Graph1.add_node(i)

list(Graph1.nodes)

#for loop through all pair of nodes and generate a random possibiliy between 0 and 1. if random number < p, have an edge. 
for i in range(n):
  for j in range(i+1, n):
    random_num = random.uniform(0,1)
    if random_num <= p:
      Graph1.add_edge(i,j)

list(Graph1.edges)

#draw graph
nx.draw(Graph1, with_labels = True)
plt.savefig('Par1_random_graph.png')
```

Deliverable 2: Create a random graph with 100 > nodes, different variations of p (run 10 times) with different number of componenets. 

```python
import matplotlib.pyplot as plt
from statistics import mean

project2_n = 106
project2_p = 0.005

Graph2 = nx.Graph()
for i in range(project2_n):
  Graph2.add_node(i)

connected_component_num = []
for i in range(11):
  connected_component_num.append([])

list_counter = 0

while project2_p <= 0.055:
  for count in range(10):
    #create a new graph Graph_05 and add nodes
    Graph2 = nx.Graph()
    for i in range(project2_n):
      Graph2.add_node(i)
    #calculate probability and add edges
    for i in range(project2_n):
      for j in range(i+1, project2_n):
        random_num = random.uniform(0,1)
        if random_num <= project2_p:
          Graph2.add_edge(i,j)
    connected_component_num[list_counter].append(nx.number_connected_components(Graph2))
  list_counter = list_counter + 1
  project2_p += 0.005

for i in range(11):
  print(connected_component_num[i])

print()
print()
print()

print("Printing scatter plot x and y elements")
scatter_plotx = []
scatter_ploty = []

for i in range(11):
  for j in range(10):
    scatter_ploty.append(connected_component_num[i][j])

x_component_counter = 0.005
for i in range(11):
  for j in range(10):
    scatter_plotx.append(x_component_counter)
  x_component_counter += 0.005

print("X elements")
print(scatter_plotx)

print("Y elements")
print(scatter_ploty)

print()
print()
print()

plt.scatter(scatter_plotx, scatter_ploty)
plt.title("Number of components in 106 nodes graph with probabilities of edges")
plt.xlabel("Probabilites of having edges")
plt.ylabel("Number of components")
plt.show()

print()
print()
print()

#calculate average and print
average = []
total = 0
for i in range(11):
  average.append(mean(connected_component_num[i]))
print(average)

print()
print()
print()

average_ploty = average
average_plotx = []
x_component_counter = 0.005
for i in range(11):
  average_plotx.append(x_component_counter)
  x_component_counter += 0.005

print(average_plotx)
print(average_ploty)



plt.plot(average_plotx, average_ploty)
plt.title("Number of average components in 106 nodes graph with probabilities of edges")
plt.xlabel("Probabilites of having edges")
plt.ylabel("Average Number of components")

for i, j in zip(average_plotx, average_ploty):
   plt.text(i, j, '({:.3f}, {})'.format(i, j))

plt.show()
```

Deliverable 3: Create a random graph with 100 > nodes, different variations of p(run 10 times) with different degree of nodes. 

```python
import matplotlib.pyplot as plt
from statistics import mean

project3_n = 106
project3_p = 0.005

Graph3 = nx.Graph()
for i in range(project3_n):
  Graph3.add_node(i)

#store all average nodes degree by p with 10 random graphs
node_degree = []
for i in range(11):
  node_degree.append([])

#list for node_degree
list_counter = 0

while project3_p <= 0.055:
  #10 random graphs
  for count in range(10):
    #create a new graph Graph_05 and add nodes
    Graph3 = nx.Graph()
    for i in range(project2_n):
      Graph3.add_node(i)
    #calculate probability and add edges
    for i in range(project3_n):
      for j in range(i+1, project3_n):
        random_num = random.uniform(0,1)
        if random_num <= project3_p:
          Graph3.add_edge(i,j)
    total = 0
    for i in range(project3_n):
      #calculate average
      total = total + Graph3.degree[i]
    average = total/project3_n
    node_degree[list_counter].append(average)
  list_counter = list_counter + 1
  project3_p += 0.005

for i in range(11):
  print(node_degree[i])
    
print("Printing scatter plot x and y elements")
scatter_plotx = []
scatter_ploty = []

for i in range(11):
  for j in range(10):
    scatter_ploty.append(node_degree[i][j])

x_component_counter = 0.005
for i in range(11):
  for j in range(10):
    scatter_plotx.append(x_component_counter)
  x_component_counter += 0.005

print("X elements")
print(scatter_plotx)

print("Y elements")
print(scatter_ploty)

print()
print()
print()

plt.scatter(scatter_plotx, scatter_ploty)
plt.title("Number of degrees in 106 nodes graph with probabilities of edges")
plt.xlabel("Probabilites of having edges")
plt.ylabel("Number of degrees")
plt.show()

print()
print()
print()

#calculate average and print
average = []
total = 0
for i in range(11):
  average.append(mean(node_degree[i]))
print(average)

print()
print()
print()

average_ploty = average
average_plotx = []
x_component_counter = 0.005
for i in range(11):
  average_plotx.append(x_component_counter)
  x_component_counter += 0.005

print(average_plotx)
print(average_ploty)

plt.plot(average_plotx, average_ploty)
plt.title("Number of average node degrees in 106 nodes graph with probabilities of edges")
plt.xlabel("Probabilites of having edges")
plt.ylabel("Average Number of degrees")

for i, j in zip(average_plotx, average_ploty):
   plt.text(i, j+0.5, '({:.3f}, {})'.format(i, j))

plt.show()
```
